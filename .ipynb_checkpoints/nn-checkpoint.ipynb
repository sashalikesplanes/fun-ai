## Single Best Joke recommendation system
For this attempt we will try to use a neural network to predict which jokes a person may like.

We will also redefine the exact problem statement.

The final application to be created should ask the user to rate several jokes to understand their latent factors and then based on this understanding of the user recommend the next joke.

For now lets make the system simply predict the one joke it believes the user will enjoy the most.

We wish for the user to rate 8 jokes with idx = {7, 8, 13, 15, 16, 17, 18, 19} - the jokes that have a rating by every user. Based on the users rating of these 8 jokes we wish to recommend the best joke to the user out of the remaining 120 jokes.

The DL architecture we desire:

X - rating of the 8 gauge jokes per user

Y - ID of the highest rated joke by the user, outside the gauge dataset

With one hidden layer representing the latent factors of the jokes.

Going from X to the hidden layer gives us an estimate of the latent factors found in the user
Going from HL to Y is the mapping of the latent factors of the jokes in comparison to the latent factors found in the user

```python
from fastai.collab import *
from fastai.tabular.all import *
```

```python
path = Path.cwd()
```

load data from excel file

```python
data_raw = pd.read_excel(path/'data'/'dataset.xls', header=None, 
                         usecols=[i for i in range(1, 151)])
```

pre proccess data as 99 encodes missing ratings, add user id column, and remove the jokes which the dataset authors highlight as no longer relevant

```python
data_raw = data_raw.replace([99.], np.nan)
data_raw["user_id"] = [i for i in range(0, 54905)]
data_raw = data_raw.drop([1, 2, 3, 4, 5, 6, 9, 10, 11, 12, 14, 20, 27, 31, 43, 51, 52, 61, 73, 80, 100, 116], axis=1)
```

```python
data_raw
```

As all the users have rated the 8 gauge jokes, we will select only the users which have rated at least 20 other jokes outside the gauge dataset, so we select the users with at least 29 entires per row, as 8 entries are gauge jokes, 1 is the user_id itself leaving 20 jokes outside the gauge dataset which they have rated. Later we will investigate adding the users which have rated less jokes.

```python
#select the 20k of users with at least 20 jokes rated
```

```python
data_raw.count(1).sort_values(ascending=False)[0:5000]
```

```python
data_raw = data_raw.loc[data_raw.count(1).sort_values(ascending=False)[0:20000].index]
```

```python
data_raw
```

```python
x_data_raw = data_raw.loc[:, [7, 8, 13, 15, 16, 17, 18, 19]]
x_data_raw.loc[23338]
```

We should also normalize this data to account for some users rating the jokes higher and lower

```python
def normalize(s):
    return (s - s.mean())/s.std()
```

```python
x_data = x_data_raw.apply(normalize, axis=1).dropna()
x_data
```

Take note that here we have changed the joke ids to be from 0 to 119 for the non gauge set

```python
y_data_raw = data_raw.drop([7, 8, 13, 15, 16, 17, 18, 19, 'user_id'], axis=1)
#y_data_raw.columns = [i for i in range(120)]
y_data = pd.DataFrame(y_data_raw.idxmax(axis=1))
y_data
```

```python
data_final = x_data.join(y_data)
data_final
```

```python
jokes_dls = TabularDataLoaders.from_df(data_final, cont_names=[7, 8, 13, 15, 16, 17, 18, 19],
                                       y_names=0, y_block=CategoryBlock(),
                                       procs=Categorify,
                                       splits=RandomSplitter(seed=42))
```

```python
jokes_dls.show_batch()
```

```python
jokes_learner = tabular_learner(jokes_dls, layers=[256, 128], metrics=error_rate, wd=0.25).to_fp16()
```

```python
jokes_learner.lr_find()
```

```python
jokes_learner.fit_one_cycle(10, 2e-1)
```

```python
jokes_learner.lr_find()
```

```python
jokes_learner.recorder.plot_loss()
```

```python
jokes_dls.valid
```

```python
data_alt = data_raw.melt(id_vars='user_id', var_name='joke_id', value_name='rating').dropna()
```

```python
data_alt
```

```python
alt_dls = CollabDataLoaders.from_df(data_alt, user_name='user_id', item_name='joke_id',
                                    rating_name='rating', seed=42, bs=128)
```

```python
alt_dls.show_batch()
```

```python
alt_learner = collab_learner(alt_dls, n_factors=100, loss_func=MSELossFlat(), 
                             y_range=(-11, 11), wd=0.1).to_fp16()
```

```python
alt_learner.lr_find()
```

```python
alt_learner.fit_one_cycle(1, 1e-2)
```

```python
alt_learner.lr_find()
```

```python
alt_learner.fit_one_cycle(2, 1e-3)
```

```python
alt_learner.lr_find()
```

```python
alt_learner.fit_one_cycle(2, 2e-4)
```
